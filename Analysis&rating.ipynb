```python
order=df.sort_values(by="title_year" , ascending = True)
Order=order[order.title_year>=2000]
Order
```

```python
df=Order.groupby(['title_year'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount'})
ax = plt.axes()
sns.set(rc={'figure.figsize':(10,8)},style='darkgrid')
ax.set_title('No. of movies in each year',fontsize=15,fontweight="bold")
sns.barplot(x=df['moviescount'], y=df['title_year'],orient='h',palette="crest",saturation=1)
for  pY,pX in enumerate(df.moviescount):
    ax.annotate(pX,xy=(pX,pY))
plt.ylabel('no.of movies',fontsize=15,fontweight='bold')
plt.xlabel('Year',fontsize=15,fontweight='bold')
plt.show()
```

```python
imp_data=Order[Order.imdb_score >= 7]
df=imp_data.groupby(['title_year'])['movie_title'].count().reset_index().rename(columns={'movie_title':'hit_movies'})
Imp_data=Order[Order.imdb_score < 7]
Imp_data=Imp_data.groupby(['title_year'])['movie_title'].count().reset_index().rename(columns={'movie_title':'flop&avg_movies'})
```

```python
f=Order.groupby(['genres'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount'})
a=Order[Order.imdb_score >= 6]
a=a.groupby(['genres'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_hit'})
b=Order[Order.imdb_score < 6]
b=b.groupby(['genres'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_flop'})
p=f.merge(b,how='left')
x=p.merge(a,how='left')
x['success_percent']=(x['moviescount_hit']/x['moviescount'])*100
x['flop_percent']=(x['moviescount_flop']/x['moviescount'])*100
x=x.fillna(0)
x.sort_values(by='moviescount', ascending=False)
```

```python
e=Order.groupby(['actor_1_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount'})
c=Order[Order.imdb_score >= 6]
c=c.groupby(['actor_1_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_hit'})
d=Order[Order.imdb_score < 6]
d=d.groupby(['actor_1_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_flop'})
g=e.merge(d,how='left')
y=g.merge(c,how='left')
y['success_percent']=(y['moviescount_hit']/y['moviescount'])*100
y['flop_percent']=(y['moviescount_flop']/y['moviescount'])*100
y=y.fillna(0)
y.sort_values(by='moviescount', ascending=False)
```

```python
h=Order.groupby(['actor_2_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount'})
i=Order[Order.imdb_score >= 6]
i=i.groupby(['actor_2_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_hit'})
j=Order[Order.imdb_score< 6]
j=j.groupby(['actor_2_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_flop'})
k=h.merge(j,how='left')
l=k.merge(i,how='left')
l['success_percent']=(l['moviescount_hit']/l['moviescount'])*100
l['flop_percent']=(l['moviescount_flop']/l['moviescount'])*100
l=l.fillna(0)
l.sort_values(by='moviescount', ascending=False)
```

```python
o=Order.groupby(['actor_3_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount'})
m=Order[Order.imdb_score >= 6]
m=m.groupby(['actor_3_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_hit'})
n=Order[Order.imdb_score < 6]
n=n.groupby(['actor_3_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_flop'})
q=o.merge(n,how='left')
r=q.merge(m,how='left')
r['success_percent']=(r['moviescount_hit']/r['moviescount'])*100
r['flop_percent']=(r['moviescount_flop']/r['moviescount'])*100
r=r.fillna(0)
r.sort_values(by='moviescount', ascending=False)
```

```python
t=Order.groupby(['director_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount'})
s=Order[Order.imdb_score >= 6]
s=s.groupby(['director_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_hit'})
u=Order[Order.imdb_score < 6]
u=u.groupby(['director_name'])['movie_title'].count().reset_index().rename(columns={'movie_title':'moviescount_flop'})
v=t.merge(u,how='left')
w=v.merge(s,how='left')
w['success_percent']=(w['moviescount_hit']/w['moviescount'])*100
w['flop_percent']=(w['moviescount_flop']/w['moviescount'])*100
w=w.fillna(0)
w.sort_values(by='moviescount', ascending=False)
```

```python
title=input('Enter title name: ')
# genre=input('Enter genre: ')
actor1=input('Enter Actor 1 Name: ')
actor2=input('Enter Actor 2 Name: ')
actor3=input('Enter Actor 3 Name: ')
director=input('Enter Director Name:')

if (y['actor_1_name'] == actor1).bool:
     y_val =y.loc[y['actor_1_name']==actor1]['success_percent']
if (l['actor_2_name'] == actor2).bool:
     l_val = l.loc[l['actor_2_name']==actor2]['success_percent']
if (r['actor_3_name'] == actor3).bool:
     r_val =  r.loc[r['actor_3_name']==actor3]['success_percent']
if (w['director_name']== director).bool:
     w_val = w.loc[w['director_name']==director]['success_percent']
    
y_val = y_val.to_list()
l_val = l_val.to_list()
r_val = r_val.to_list()
w_val= w_val.to_list()
val = 0.25*(l_val[0]+y_val[0]+r_val[0]+w_val[0])
val
if val>=50:
    print("Rating of "+title+" lies in between 6.5 to 10")
else:
    print("Rating of "+title+" will lie below 6.5")    
```
